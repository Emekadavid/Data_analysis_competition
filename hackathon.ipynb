## Problem Statement

Africa is facing a new health crises. This problem has to be solved otherwise the huge amount of money spent by African elites and middle class on medical tourism will continue to increase. This takes away from the resources the continent needs to develop. Also, if this problem is not solved, it will continue to lose its best brains to developed countries. The highly skilled personnel are supposed to contribute to the growth and development of their country, thereby increasing its GDP. 

African governments, from Johannessburg to Cairo, need to take proactive measures to address these issues. Some of the issues involves Africans dying from diseases that are easily cured in other countries, or can be prevented when there are good health facilities and doctors that can serve the population. African population is growing at a rapid rate. If its best medical personnel increasingly travel abroad for greener pastures, there is no way Africa can sustainably address these issues. 

These data analytic process seeks to examine what the data says about the health crises presently facing African countries. Using the data, it will seek to profer solutions to these health crises and also outline a vision for Africans and African governments.

### Question 3: Are the number of medical doctors in Africa sufficient?

According to a news bulleting cited in the footnotes [3], the ideal doctor-population ratio is 1:1000. That means ideally, a doctor should be catering to 1000 patients at any point in time. 

Let us see what the data says about African doctor-population ratio and then find out if we have sufficient doctors in Africa.

```python
# importing an excel file. 
doctor_pop_df = pd.read_excel("3. Medical Doctors Per 10000 population.xlsx", header=2)
doctor_pop_df.head()
```

We need to bring in the population data per country. This will help us do the analysis proper. 

```python
population_df = pd.read_csv("5. World Population.csv")
population_df.head()
```

We can see some negative values in the year column. Let's investigate that column further. 

First, we'll find out the unique years in the column. 

```python
population_df.Year.unique()
```

we can see that there are several population data. We'll use the year, 2018 because that is the latest year in the doctor's dataset. 

So we filter the population table by year. 

```python
population_df = population_df[population_df["Year"] == 2018]
population_df.head()
```

We can now see from the Entity column (representing countries), there is a mix of both countries and continents. Let's investigate that further. 

```python
# the unique values in the Entity column
population_df.Entity.unique()
```

So, we can see that there is a mixture of country and continent data. This analysis is interested in only the population for Africa and European continent, and the countries in them. We'll program the populations into the doctor's dataframe. This will then be useful for our purposes. North America is not used this time because that region is not characterized so in this data set. 

To start, we need to filter the doctor's column by a corresponding year. 

```python
doctor_pop_df["Period"].unique()
```

The latest year is 2018, so we filter by 2018 so it will align with the population data set. 

```python
doctor_pop_df = doctor_pop_df[doctor_pop_df["Period"] == 2018]
doctor_pop_df.head()
```

we can then align the populations that will be useful for calculating the doctor:patient ratio. 

We need to merge the two data sets. But first we convert the value column to numeric

```python
doctor_pop_df.dtypes
```

```python
def get_population(x):
    series_name = population_df[population_df["Entity"] == x]
    population = series_name["Population (historical estimates)"]
    population = population.values
    if len(population) == 0:
        population = 0
        return population
    else:
        population = population[0]
        return population   
```

```python
doctor_pop_df["Population"] = doctor_pop_df["Location"].apply(get_population)
```

```python
doctor_pop_df.head()
```

We create a number of doctors estimate for each country.  

```python
def number_of_doctors(x):
    # value * population divided by 10000 gives number of doctors estimate for that country
    number = (float(x["FactValueNumeric"]) * float(x["Population"]))/10000
    number = round(number, 0)
    return number
```

```python
doctor_pop_df["total_doctors"] = doctor_pop_df.apply(number_of_doctors, axis=1)
doctor_pop_df.head()
```

We now filter by Africa and get the ratio. Then filter by Europe and get the ratio.

```python
african_doctor_ratio = doctor_pop_df[doctor_pop_df["ParentLocation"] == "Africa"]
african_doctor_ratio.head()
```

```python
total_doctors =  sum(african_doctor_ratio["total_doctors"])
total_population = sum(african_doctor_ratio["Population"])
print(f'There are {total_doctors} doctors in Africa as of 2018 and {total_population} people')
doctor_ratio = total_doctors/total_population
print(f'That means in African 1 doctor serves {doctor_ratio} persons')
```

```python
european_doctor_ratio = doctor_pop_df[doctor_pop_df["ParentLocation"] == "Europe"]
european_doctor_ratio.head()
```

```python
total_doctors =  sum(european_doctor_ratio["total_doctors"])
total_population = sum(european_doctor_ratio["Population"])
print(f'There are {total_doctors} doctors in Europe as of 2018 and {total_population} people')
doctor_ratio = total_doctors/total_population
print(f'That means in Europe 1 doctor serves {doctor_ratio} persons')
```

These results shows that because Africa has a huge population, there are more doctors in Africa than in developed continents like Europe. Also, the doctor to patient ratio is low than in Europe.  


Let's compare the mean doctors to 10,000 persons ratio for further insight. 

```python
africa_10k_mean = african_doctor_ratio["FactValueNumeric"].mean()
europe_10k_mean = european_doctor_ratio["FactValueNumeric"].mean()
print(f'mean doctor to 10k population ratio for Africa is {africa_10k_mean}')
print(f'mean doctor to 10k population ratio for Europe is {europe_10k_mean}')
```

So the mean of doctors to 10,000 population confirms it that there are several doctors in Africa compared to other places like Europe.
